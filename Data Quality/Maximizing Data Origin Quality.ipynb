# Data Validation 

### Now we want to know the validition of qulity for this data that we were worked with, therefore we have A Regression model might be better at generalizing to unseen data so we will use Random Forest Regression (RFR) Algorithm with SKlearn

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Select relevant columns for visualization and prediction
features = ['discounted_price', 'rating', 'rating_count']
target = 'actual_price'
```

```python
# Visualize the dataset
sns.pairplot(df[features + [target]])
plt.show()
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
```

```python
# Create a Random Forest Regression model
model = RandomForestRegressor()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```

### Visualize the predicted and actual prices

```python

plt.scatter(X_test['discounted_price'], y_test, color='blue', label='Actual Price')
plt.scatter(X_test['discounted_price'], y_pred, color='red', label='Predicted Price')
plt.xlabel('Discounted Price')
plt.ylabel('Actual/Predicted Price')
plt.legend()
plt.show()
```
