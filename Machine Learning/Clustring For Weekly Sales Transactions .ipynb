# Weekly Sales Transactions 

### Import necessary library and Read the dataset

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


df = pd.read_csv("Sales_Transactions.csv")
```

### Drop the Product_Code column

```python
df = df.drop("Product_Code", axis=1)
```

### Normalize the data

```python
df = df.apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

### Create a KMeans object with 5 clusters

```python
kmeans = KMeans(n_clusters=5)
```

### Fit the KMeans model to the data

```python
kmeans.fit(df)
```

### Predict the cluster labels for each data point and Print it

```python
labels = kmeans.predict(df)

print(labels)
```

### Visualize the clusters

```python
plt.scatter(df["Normalized 0"], df["Normalized 1"], c=labels, alpha=0.5)
plt.xlabel("Normalized 0")
plt.ylabel("Normalized 1")
plt.title("KMeans Clustering")
plt.show()
```
